```python
#Packages that we'll use during process
from snowballstemmer import TurkishStemmer
import os #for accessing os path
import docx2txt #this package converts our docx files to txt
import pandas as pd
import itertools
import numpy as np
import re
turkish_stemmer = TurkishStemmer()
#Change direction according to your OS
os.chdir(r'/Users/burakhasici/Desktop/turkish-spam-filter')
path = r"/Users/burakhasici/Desktop/turkish-spam-filter"
```

```python
#Library that we'll use for word tokenazation.
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
files = []
text_list = []
all_list = []
exc_mark_count_list = []
for file in os.listdir(path):
    if file.endswith('.docx'):
        files.append(file)
for i in range(len(files)):
    text = docx2txt.process(files[i])
    text_list = word_tokenize(text)
    #text_df = pd.DataFrame(text_list)
    #turkish_stemmed_words = turkish_stemmer.stemWords(text_list)
    all_list.append(text_list)
#Turned our array a numpy array for better list manipulation
np_all_list = np.array(all_list,dtype="object")
print(np_all_list)
```
