```python
files = []
text_list = []
all_list = []
non_tokenized_list = []
exc_mark_count_list = []
for file in os.listdir(path):
    if file.endswith('.docx'):
        files.append(file)
for i in range(len(files)):
    text = docx2txt.process(files[i])
    non_tokenized_list.append(text)
    text_list = word_tokenize(text)
    all_list.append(text_list)
#Turned our array a numpy array for better list manipulation
np_all_list = np.array(all_list,dtype="object")
```

```python
turkish_lira_list = []
turkish_lira_possiblilities = ["türk","lirası","lira","tl","₺","turk","lıra"]

def count_turkish_liras(mail_list):
    for i in range(len(mail_list)):
        for j in range(len(mail_list[i])):
            mail_list[i][j] = mail_list[i][j].lower()
    for i in range(len(mail_list)):
        count = len(set.intersection(set(turkish_lira_possiblilities),set(mail_list[i])))
        turkish_lira_list.append(count)
    #print(spam_word_count_list)
    return turkish_lira_list
```

```python
dollar_list = []
dollar_possibilities = ["dolar","dollar","$"]
def count_dollar(mail_list):
    for i in range(len(mail_list)):
        for j in range(len(mail_list[i])):
            mail_list[i][j] = mail_list[i][j].lower()
    for i in range(len(mail_list)):
        count = len(set.intersection(set(dollar_possibilities),set(mail_list[i])))
        dollar_list.append(count)
    #print(spam_word_count_list)
    return dollar_list
```

```python
spam_word_count_list = []
total_spam_words_in_spam_mails = []
def count_spam_words(mail_list):
    for i in range(len(mail_list)):
        for j in range(len(mail_list[i])):
            mail_list[i][j] = mail_list[i][j].lower()
    for i in range(len(mail_list)):
        count = len(set.intersection(set(spam_words),set(mail_list[i])))
        spam_word_count_list.append(count)
    #print(spam_word_count_list)
    return spam_word_count_list

total_spam_words_in_spam_mails.append(count_spam_words(np_all_list))
print(total_spam_words_in_spam_mails)
```

```python
from nltk import ngrams
grams_list = []
grams_tuple_list = []
grams_tuple_concencanate_list = []
n = 2

def join_tuple_string(strings_tuple) -> str:
    return ' '.join(strings_tuple)

def n_gram_convertor(mail_list):
    for i in range(len(mail_list)):
        twograms = ngrams(mail_list[i],n)
        grams_list.append(twograms)
        for grams in grams_list[i]:
            b = [i for sub in grams_list for i in sub]
            grams_tuple_list.append(b)
    for i in range(len(grams_tuple_list)):
        result = list(map(join_tuple_string,grams_tuple_list[i]))
        grams_tuple_concencanate_list.append(result)
    return grams_tuple_concencanate_list

print(n_gram_convertor(np_all_list))
```
